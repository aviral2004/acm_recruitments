```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pickle as pkl
```

```python
with open('./test_data/layer_depth.pkl', 'rb') as f:
    depth_data = pkl.load(f)

with open('./test_data/layer_width.pkl', 'rb') as f:
    width_data = pkl.load(f)
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 7))

ax[0].set_ylabel("loss")
ax[0].set_xlabel("epochs")
for i in range(len(depth_data)):
    ax[0].plot(depth_data[i]['loss_array'], label = f"depth = {depth_data[i]['layers']}")

ax[1].set_ylabel("accuracy")
ax[1].set_xlabel("epochs")
for i in range(len(depth_data)):
    ax[1].plot(depth_data[i]['accuracy_array'], label = f"depth = {depth_data[i]['layers']}")

ax[0].set_title("loss vs. epochs while changing NN depth")
ax[1].set_title("accuracy vs. epochs while changing NN depth")


f.tight_layout()
ax[0].legend()
ax[1].legend()
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))

ax[0].set_ylabel("loss")
ax[0].set_xlabel("epochs")
for i in range(len(depth_data)):
    ax[0].plot(width_data[i]['test_loss_array'], label = f"width = {width_data[i]['nodes']}")
    # ax[0].plot(width_data[i]['train_loss_array'], label = f"width = {width_data[i]['nodes']} (train)")


ax[1].set_ylabel("accuracy")
ax[1].set_xlabel("epochs")
for i in range(len(depth_data)):
    ax[1].plot(width_data[i]['accuracy_array'], label = f"width = {width_data[i]['nodes']}")

ax[0].set_title("loss vs. epochs while changing NN layer width")
ax[1].set_title("accuracy vs. epochs while changing NN layer width")

f.tight_layout()
ax[0].legend()
ax[1].legend()
```
